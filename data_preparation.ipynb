#### Import Required Libraries

```python
import pandas as pd
```

```python
data = pd.read_csv('data\part_1\historical_weather.csv')
data["date"] = pd.to_datetime(data["date"])
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data = data[["date", "city_id", "avg_temp_c"]]
data
```

```python
# Get the number of null values in the avg_temp_c column for each city
data[data["avg_temp_c"].isnull()].groupby("city_id").size()
```

```python
def dates_with_nulls(city):
    return data[(data["city_id"]==city) & (data["avg_temp_c"].isnull())].reset_index(drop=True)
```

```python
dates_with_nulls("C018")
```

```python
dates_with_nulls("C027")
```

```python
dates_with_nulls("C082")
```

```python
dates_with_nulls("C090")
```

```python
dates_with_nulls("C094")
```
