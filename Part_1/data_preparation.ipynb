##### Replacing Null Values

```python
# Make a function for filling the null values in the avg_temp_c column
def fill_nulls(city):
    city_data = data[data["city_id"]==city].reset_index(drop=True)
    nulls = city_data[city_data["avg_temp_c"].isnull()].reset_index(drop=True)
    for i in range(len(nulls)):
        date = nulls["date"][i]
        month = date.month
        day = date.day
        avg_temp = round(city_data[city_data["date"].dt.month==month][city_data["date"].dt.day==day]["avg_temp_c"].mean(), 1)
        data.loc[(data["city_id"]==city) & (data["date"]==date), "avg_temp_c"] = avg_temp
        print(f"Filled the null value in the avg_temp_c column for city {city} at date {date} with the value {avg_temp}")
```

The above null values remain due to 2016 being a leap year, which cannot be replaced with the mean of other months. Therefore, we can use forward fill to handle these null values. 

```python
# Forward fill the remaining null values
data = data.fillna(method="ffill")
```

### Handling Outliers

```python
def handle_outliers(group):
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = group['avg_temp_c'].quantile(0.25)
    Q3 = group['avg_temp_c'].quantile(0.75)
    IQR = Q3 - Q1
    
    # Calculate the lower and upper bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Replace outliers with the median of the city
    median = group['avg_temp_c'].median()
    group['avg_temp_c'] = group['avg_temp_c'].apply(lambda x: median if x < lower_bound or x > upper_bound else x)
    return group

# Apply the function to each city
data_cleaned = data.groupby('city_id').apply(handle_outliers)

data_cleaned.head(20)
```

```python
# Save data without outliers to a new CSV file
data_cleaned.to_csv(r"..\data\part_1\historical_weather_without_outliers.csv", index=False)
```
