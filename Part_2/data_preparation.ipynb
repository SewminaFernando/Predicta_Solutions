```python
import pandas as pd
```

```python
data = pd.read_csv(r"..\data\part_2\daily_data.csv")
data
```

```python
# Check null values
data.isnull().sum()
```

```python
data = data.dropna()
data.reset_index(drop=True, inplace=True)
data
```

```python
data["condition_text"].value_counts()
```

```python
# Check for duplicates
data.duplicated().sum()
```

```python
# Drop sunrise and sunset columns
data.drop(columns=["sunrise", "sunset"], inplace=True)
```

```python
# get condition_text column to the right corner
condition_text = data.pop("condition_text")
data["condition_text"] = condition_text
```
