```python
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Load the updated dataset
data = pd.read_csv('data/part_1/historical_weather_without_outliers.csv')

# Convert the 'date' column to datetime format
data['date'] = pd.to_datetime(data['date'])

# Set the 'date' column as the index
data.set_index('date', inplace=True)

# Function to train a SARIMA model for a given city
def train_sarima_model(city_data):
    # Define the SARIMA model
    model = SARIMAX(city_data['avg_temp_c'], order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
    
    # Fit the model
    model_fit = model.fit()
    
    return model_fit

# Dictionary to store SARIMA models for each city
sarima_models = {}

# Train a SARIMA model for each city
for city_id, city_data in data.groupby('city_id'):
    sarima_models[city_id] = train_sarima_model(city_data)

# Function to forecast using the trained SARIMA model
def forecast_sarima_model(model, steps):
    forecast = model.forecast(steps=steps)
    return forecast

# Dictionary to store forecasted values for each city
sarima_forecasts = {}

# Forecast the first week of 2019 for each city and print the forecasted values
for city_id, model in sarima_models.items():
    forecast = forecast_sarima_model(model, steps=7)
    sarima_forecasts[city_id] = forecast
    print(f"Forecasted values for city {city_id}:")
    print(forecast)
    print("\n")

# Save the forecasted values to a CSV file
forecast_data = []
submission_id = 1
for city_id, forecast in sarima_forecasts.items():
    for temp in forecast:
        forecast_data.append({'submission_ID': submission_id, 'avg_temp_c': round(temp, 2)})
        submission_id += 1
```

```python
forecast_df = pd.DataFrame(forecast_data)
forecast_df.to_csv('data/part_1/forcasted_data/sarima.csv', index=False)

print("Forecast data has been saved to 'sarima.csv'")
```
